```python
import pandas as pd
import numpy as np
```

```python
# Marvel Comics, Lucasfilm, Pixar, Walt Disney Animation Studios, Blue Sky Studios, Disney Channel, DisneyToon Studios, Disneynature
marvel = pd.read_csv('marvel_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
lucas = pd.read_csv('lucasfilm_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
pixar = pd.read_csv('pixar_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
disney_animation = pd.read_csv('disney_animation_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
blue_sky = pd.read_csv('blue_sky_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
disney_channel = pd.read_csv('disney_channel_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
disneytoon = pd.read_csv('disneytoon_movies.csv', usecols=lambda column: column != 'Unnamed: 0')
disneynature = pd.read_csv('disneynature_movies.csv', usecols=lambda column: column != 'Unnamed: 0')

marvel['Brand'] = 'Marvel Comics'
lucas['Brand'] = 'Lucasfilm'
pixar['Brand'] = 'Pixar'
disney_animation['Brand'] = 'Walt Disney Animation Studios'
blue_sky['Brand'] = 'Blue Sky Studios'
disney_channel['Brand'] = 'Disney Channel'
disneytoon['Brand'] = 'DisneyToon Studios'
disneynature['Brand'] = 'Disneynature'

disney = pd.concat([marvel, lucas, pixar, disney_animation, blue_sky, disney_channel, disneytoon, disneynature], ignore_index=True)
disney
```

```python
def format_gross_income(value):
    return f"${value:,.0f}"
```

```python
disney_sorted = disney.copy()
disney_sorted['Gross Income'] = pd.to_numeric(disney_sorted['Gross Income'].str.replace('[$,]', '', regex=True))
disney_sorted = disney_sorted.sort_values(by='Gross Income', ascending=False).reset_index(drop = True)
disney_sorted['Rank'] = disney_sorted.index + 1
disney_sorted.rename(columns={'Rank': 'New Rank'}, inplace=True)
disney_sorted['Gross Income'] = disney_sorted['Gross Income'].apply(format_gross_income)
print(disney_sorted.shape[0])
disney_sorted.head(15)
```

```python
disney_sorted.to_csv("disney_owned_movies.csv")
```
