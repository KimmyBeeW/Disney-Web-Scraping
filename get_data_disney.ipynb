```python
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from RequestGuard import RequestGuard as RG
```

Downfall of Disney

```python
load_dotenv('.env')
user_agent = os.environ.get('USER_AGENT')
header = {'User-Agent': user_agent}
```

```python

url2 = 'https://finance.yahoo.com/quote/DIS/history/?period1=1571423893&period2=1729276688'
rg = RG(url2)
if rg.can_follow_link(url2):
    print("robot.txt allows scraping for this page")
else:
    print('Forbidden: ', rg.forbidden)
```

```python
r2 = requests.get(url2, headers=header)
print(r2.status_code)
```

```python
soup2 = BeautifulSoup(r2.text, 'html.parser')
# print(soup2.title.text)
container2 = soup2.find('div', {'class': "table-container yf-ewueuo"})
container2
```

```python
table = container2.find('table', {'class': 'yf-ewueuo'})
# table
items = table.find_all('tr', {'class': 'yf-ewueuo'})
print(items[0])
print(items[1])
```

```python
headers = [th.get_text(strip=True) for th in table.find('tr').find_all('th')]
headers[4] = 'Close'
headers[5] = 'Adj Close'
print(headers)

rows = []
for tr in items[1:]:
    row = [td.get_text(strip=True) for td in tr.find_all('td')]
    rows.append(row)
print(rows)
```

```python
dis_df = pd.DataFrame(rows, columns=headers)
dis_df
```

```python
dis_df.to_csv('disney_stocks.csv')
```
